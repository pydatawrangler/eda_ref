# Chapter 13 - Dates in the Index

```python
snow = (alta_df
    .SNOW
    .rename(dates)
)

snow
```

```python
snow.isna().any()
```

```python
snow.isna().sum()
```

```python
snow[snow.isna()]
```

```python
snow.loc['1985-09':'1985-09-28']
```

```python
(snow
    .loc['1985-09':'1985-09-20']
    .fillna(0)
)
```

```python
snow.loc['1987-12-30':'1988-01-10']
```

```python
(snow
 .loc['1987-12-30':'1988-01-10']
 .ffill()
 )
```

```python
(snow
 .loc['1987-12-30':'1988-01-10']
 .bfill()
 )
```

## 13.3 Interpolation

```python
(snow
 .loc['1987-12-30':'1988-01-10']
 .interpolate()
 )
```

```python
winter = (snow.index.quarter == 1) | (snow.index.quarter == 4)
```

```python
(snow
 .where(~(winter & snow.isna()), snow.interpolate())
 .where(~(~winter & snow.isna()), 0)
 )
```

```python
(snow
 .where(~(winter & snow.isna()), snow.interpolate())
 .where(~(~winter & snow.isna()), 0)
 .loc[['1985-09-19', '1988-01-01']]
 )
```
