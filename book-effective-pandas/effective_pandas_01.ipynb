```python
top5 = make.value_counts().index[:5]
```

```python
def generalize_top5(val):
    if val in top5:
        return val
    return 'Other'
```

## 11.5 Splitting

```python
age = pd.Series(['0-10','11-15','11-15','61-65','46-50'])
age
```

```python
age.str.split('-')
```

```python
(age
 .str.split('-', expand=True)
 .iloc[:,0]
 .astype(int)
)
```

```python
(age
 .str.slice(-2)
 .astype(int)
 )
```

```python
(age
 .str[-2:]
 .astype(int)
 )
```

```python
(age
 .str.split('-', expand=True)
 .astype(int)
 .mean(axis='columns')
 )
```

```python
import random
```

```python
def between(row):
    return random.randint(*row.values)

def star_rows(row):
    print(*row.values)
```

```python
(age
 .str.split('-', expand=True)
 .astype(int)
 .apply(between, axis='columns')
 )
```

```python
(age
 .str.split('-', expand=True)
 .astype(int)
 .apply(star_rows, axis='columns')
 )
```

## 11.6 Optimizing `.apply` with Cython

```python
%load_ext Cython
```

```python
import random
```

```python
%%cython
import random
def between_cy(row):
    return random.randint(*row.values)
```

```python
%%cython
import random
cpdef int between_cy3(int x, int y):
    return random.randint(x, y)
```

```python
%prun -l 10 (age.str.split('-', expand=True).astype(int).apply(lambda row: between_cy3(row[0], row[1]), axis=1))
```
